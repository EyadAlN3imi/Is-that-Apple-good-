```python
data.drop(['Crunchiness','Weight'],axis=1,inplace=True)
```

```python
data.info()
```

```python
x = data.drop('Quality',axis=1)
y = data['Quality']
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
Rand_Forest = RandomForestClassifier(max_depth=25,min_samples_leaf=1,min_samples_split=2,n_estimators=100,n_jobs=-1)
Rand_Forest.fit(x_train,y_train)
Rand_pred = Rand_Forest.predict(x_test)
print(f"RF Accuracy: {accuracy_score(y_test,Rand_pred):0.2%}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
conf = confusion_matrix(y_test, Rand_pred)
plt.title("Random Forest ConfusionMatrix")
a = sns.heatmap(conf,annot=True,fmt='g')
a.set_xticklabels(['good', 'bad'])
a.set_yticklabels(['good', 'bad'])
plt.show()
```

That's not all we will improve the accuracy by Ahmad Alaam technique

```python
accuracy_score(Rand_Forest.predict(x),y)
```

```python
data = pd.concat([data,pd.DataFrame(Rand_Forest.predict(x))],axis=1)
```

```python
data.rename(columns={0: 'Predicted'}, inplace=True)
```

```python
data
```

```python
wrong = data[data['Quality'] != data['Predicted']]
right = data[data['Quality'] == data['Predicted']]
```

```python
wrong["FPN"] = "F"
right["TPN"] = "T"
```

```python
print(wrong.info())
print(right.info())
```

```python
union = pd.concat([right, wrong])
```

```python
union.info()
```

```python
union.drop('Predicted',axis=1,inplace=True)
```

```python
union['TFPN'] = union['TPN'].fillna('') +  union['FPN'].fillna('')
union.drop(['TPN',"FPN"],axis=1,inplace=True)
```

```python
union
```

```python
xnew = union.drop(['TFPN','Quality'],axis=1)
ynew = union['TFPN']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(xnew,ynew,test_size=0.20,random_state=42)
```

```python
rf = RandomForestClassifier(max_depth=25,min_samples_leaf=1,min_samples_split=2,n_estimators=100,n_jobs=-1).fit(xtrain,ytrain)
rf_pred = rf.predict(xtest)
accuracy_score(rf_pred,ytest)
```

ACTUALLY HERE WE WILL START THE MODEL ALL BEFORE WAS JUST TO USE THE ERROR AS A CLASS To HELP US IMPROVING THE ACCURACY !!!!!!!!!!!

```python
df = pd.concat([x,pd.DataFrame(rf.predict(x)),y],axis=1)
df.rename(columns={0:"TF"},inplace=True)
df
```

```python
df = pd.DataFrame(pd.concat([pd.get_dummies(df["TF"]),df],axis=1).drop("TF",axis=1))
df
```

```python
X = df.drop('Quality',axis=1)
Y = df['Quality']
xt,xs,yt,ys = train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
Random_Forest = RandomForestClassifier(max_depth=25,min_samples_leaf=1,min_samples_split=2,n_estimators=100,n_jobs=-1).fit(xt,yt)
Random_Forest_predictd = Random_Forest.predict(xs)
print(f"RF Accuracy: {accuracy_score(ys,Random_Forest_predictd):0.2%}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
con = confusion_matrix(Random_Forest_predictd,ys)
plt.title("Random Forest ConfusionMatrix")
a = sns.heatmap(conf,annot=True,fmt='g')
a.set_xticklabels(['good', 'bad'])
a.set_yticklabels(['good', 'bad'])
plt.show()
```
